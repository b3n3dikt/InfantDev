# QC Image Gallery
This notebook generates an HTML file to display QC images for each subject.

```python
import os
from IPython.display import display, HTML

# Read the subject/session list from a text file
subjects_sessions_file = '/home/bramirez/projects/InfantDevelopment/NKIdev/info/subs_and_sessions.txt'
#subject_list_path = "/home/bramirez/projects/InfantDevelopment/NKIdev/info/sub-XX_ses-XX_list.txt"
#qc_folder = "/projects/NHP_processed/developmental_out/QC_images/anat_processed_QC_images_t2"
#output_file = "/projects/NHP_processed/developmental_out/QC_images/anat_processed_QC_images/QC_images.html"


with open(subjects_sessions_file, 'r') as f:
    subjects_sessions = [line.strip() for line in f.readlines()]

# Customize the base path
base_path = '/projects/NHP_processed/developmental_out/'

html_template = '''
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Quality Control</title>
    <style>
        body {{
            font-family: Arial, sans-serif;
        }}
        .subject {{
            margin-bottom: 30px;
        }}
        .subject h2 {{
            margin-bottom: 10px;
        }}
        .subject img {{
            width: 100%;
            max-width: 400px;
            margin-bottom: 10px;
        }}
    </style>
</head>
<body>
    <h1>Quality Control Images</h1>
    {content}
</body>
</html>
'''

subject_template = '''
<div class="subject">
    <h2>Subject: SUB-{subject_id} / Session: SES-{session_id}</h2>
    {images}
</div>
'''

image_template = '<img src="{src}" alt="Subject {subject_id} - Image {image_number}">\n'

content = ''

for subject_session in subjects_sessions:
    subject_id, session_id = subject_session.split('/')[0][-3:], subject_session.split('/')[1][-3:]
    images = ''
    image_paths = [
        f'{base_path}QC_images/anat_processed_QC_images_t2/sub-{subject_id}_ses-{session_id}_aseg_acpc_QC.png',
        f'{base_path}QC_images/anat_processed_QC_images_t2/sub-{subject_id}_ses-{session_id}_surfaces_wBrain_QC.png',
        f'{base_path}QC_images/mask_QC_images/sub-{subject_id}_ses-{session_id}_all_views_wide_format_mask_qc.png',
        f'{base_path}sub-{subject_id}/ses-{session_id}/files/summary_DCANBOLDProc_v4.0.0/DVARS_and_FD_CONCA_ses-{session_id}_task-REST.png',
        f'{base_path}sub-{subject_id}/ses-{session_id}/files/summary_DCANBOLDProc_v4.0.0/DVARS_and_FD_CONCP_ses-{session_id}_task-REST.png'
    ]

    for i, path in enumerate(image_paths):
        if os.path.exists(path):
            images += image_template.format(src=path, subject_id=subject_id, image_number=i + 1)

    if images:
        content += subject_template.format(subject_id=subject_id, session_id=session_id, images=images)

html_output = html_template.format(content=content)

# Write the HTML output to a file
with open('qc_bids_format.html', 'w') as f:
    f.write(html_output)
```
